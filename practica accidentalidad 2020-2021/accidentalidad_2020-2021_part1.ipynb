## Lets work the date and hour columns

```python
#let's clasify the hour column into 3 groups 
# 1. high: from 7:30 to 9:30 and from 18:00 to 20:00 monday to friday
# 2. night: from 0:00 to 7:30
# 3. low: the rest

#for this, we need to know de day of the week (from the date) 
# and we need to convert the "fecha" and "hora" columns to date_time

df['fecha']= pd.to_datetime(df['fecha'])
df['hora']= pd.to_datetime(df['hora'])
test['fecha']= pd.to_datetime(test['fecha'])
test['hora']= pd.to_datetime(test['hora'])
df.info()
```

```python
# we will extract the day of the week (dow) from the date
df_dow=df["fecha"].apply(lambda fila: fila.weekday())
df.insert(2, "dow", df_dow)
test_dow=test["fecha"].apply(lambda fila: fila.weekday())
test.insert(2, "dow", test_dow)
```

```python
#let's clasify the hour column into 3 groups 
# 1. high: from 7:30 to 9:30 and from 18:00 to 20:00 monday to friday
# 2. night: from 0:00 to 7:30
# 3. low: the rest
time_interval_df=df.apply(lambda dfr: 'high' if (dfr['dow']>=0) & (dfr['dow']<5) & ((pd.Timestamp('7:30') <= dfr['hora']) & (pd.Timestamp('9:30') > dfr['hora']) | (pd.Timestamp('18:00') <= dfr['hora'])  & (pd.Timestamp('20:00') > dfr['hora'])) else "night" if (pd.Timestamp('00:00') <= dfr['hora'])  & (pd.Timestamp('7:30') > dfr['hora']) else "low", axis=1)
df.insert(3, "time_int", time_interval_df)
time_interval_test=test.apply(lambda dfr: 'high' if (dfr['dow']>=0) & (dfr['dow']<5) & ((pd.Timestamp('7:30') <= dfr['hora']) & (pd.Timestamp('9:30') > dfr['hora']) | (pd.Timestamp('18:00') <= dfr['hora'])  & (pd.Timestamp('20:00') > dfr['hora'])) else "night" if (pd.Timestamp('00:00') <= dfr['hora'])  & (pd.Timestamp('7:30') > dfr['hora']) else "low", axis=1)
test.insert(3, "time_int", time_interval_test)
```

```python
# As we see in the last box, we have 26 different types of vehicle tipe. now we will see the % of each. 
aux=df1.groupby('tipo_vehiculo').count().iloc[:,0:1].sort_values(by='dow',ascending=False)
aux['%']=aux['dow']*100/aux['dow'].sum()
aux
```

```python
# 68% of the vehicles are ordinary cars, 12% are motorcycles and 8% are vans. That makes 88%. the other 32 categories, 
# correspond to the remaining 12%. So we will reorganice all vehicle-tipes into 4 categories:
# Car, Van, Motorcycle and Others
df1['grupo_vehi']=df1['tipo_vehiculo'].apply(lambda x: 'turismo' if x=='Turismo' else 'furgoneta' if x=='Furgoneta' else 'ciclomotor' if (x=='Motocicleta hasta 125cc')|(x=='Motocicleta > 125cc') else 'resto')

# And we do the same for test 
test1['grupo_vehi']=test1['tipo_vehiculo'].apply(lambda x: 'turismo' if x=='Turismo' else 'furgoneta' if x=='Furgoneta' else 'ciclomotor' if (x=='Motocicleta hasta 125cc')|(x=='Motocicleta > 125cc') else 'resto')
```

```python
df1
```

```python
# Now we will do the one_hot encoding for 'grupo_vehi' and remove the columns regarding to vehicles ('grupo_vehi' & 'tipo_vehiculo)
aux=pd.get_dummies(df1[['grupo_vehi']],columns=['grupo_vehi'],prefix='vehi_')
df2=pd.concat([df1,aux],axis=1)
df2=df2.drop(['grupo_vehi','tipo_vehiculo'], axis=1)

# Now we will do the same for test
aux_test=pd.get_dummies(test1[['grupo_vehi']],columns=['grupo_vehi'],prefix='vehi_')
test2=pd.concat([test1,aux_test],axis=1)
test2=test2.drop(['grupo_vehi','tipo_vehiculo'], axis=1)
df2

```
